```python
# CONTOH KASUS PEKERJAAN
def job_expert_system():
    """Sistem pakar sederhana untuk rekomendasi pekerjaan"""

    # Knowledge base (basis pengetahuan)
    skills_kb = {
        "programming": ["software_developer", "data_analyst"],
        "statistics": ["data_analyst"],
        "design": ["digital_marketing"],
        "communication": ["digital_marketing", "data_analyst"],
        "problem_solving": ["software_developer", "data_analyst"]
    }

    jobs_kb = {
        "software_developer": {
            "recommendation": "Apply for Software Developer position",
            "level": "Technical"
        },
        "data_analyst": {
            "recommendation": "Apply for Data Analyst position",
            "level": "Analytical"
        },
        "digital_marketing": {
            "recommendation": "Apply for Digital Marketing position",
            "level": "Creative"
        }
    }

    # Inference engine sederhana
    def recommend_job(skills):
        possible_jobs = {}
        for skill in skills:
            if skill in skills_kb:
                for job in skills_kb[skill]:
                    possible_jobs[job] = possible_jobs.get(job, 0) + 1

        # Cari pekerjaan dengan kecocokan terbanyak
        if possible_jobs:
            most_suitable = max(possible_jobs, key=possible_jobs.get)
            confidence = possible_jobs[most_suitable] / len(skills)
            return most_suitable, confidence
        return None, 0

    # Contoh penggunaan
    applicant_skills = ["programming", "problem_solving", "statistics"]
    job, confidence = recommend_job(applicant_skills)

    if job:
        recommendation = jobs_kb[job]["recommendation"]
        print(f"Recommended Job: {job} (confidence: {confidence:.0%})")
        print(f"Suggestion: {recommendation}")
    else:
        print("No suitable job found")

# Jalankan contoh
print("\n=== SISTEM PAKAR REKOMENDASI PEKERJAAN ===")
job_expert_system()
```
