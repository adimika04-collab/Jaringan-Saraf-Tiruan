```python
def job_expert_system():

    skills_kb = {
        "programming": ["software_dev", "data_analyst"],
        "statistics": ["data_analyst"],
        "communication": ["data_analyst", "marketing"],
        "problem_solving": ["software_dev"],
        "leadership": ["manager"]
    }

    jobs_kb = {
        "software_dev": "Software Developer",
        "data_analyst": "Data Analyst",
        "marketing": "Digital Marketing",
        "manager": "Project Manager"
    }

    def recommend(skills):
        result = {}
        for s in skills:
            if s in skills_kb:
                for j in skills_kb[s]:
                    result[j] = result.get(j, 0) + 1
        return result

    applicant_skills = ["programming", "statistics", "communication", "leadership"]
    results = recommend(applicant_skills)

    print("\n=== REKOMENDASI PEKERJAAN ===")
    for job, score in sorted(results.items(), key=lambda x: x[1], reverse=True):
        print(f"{jobs_kb[job]} → cocok {score} skill ({score/len(applicant_skills):.0%})")

job_expert_system()
```
