```python
class SimpleAI:
    def __init__(self, name, knowledge_base):
        self.name = name
        self.knowledge_base = knowledge_base

    def perceive(self, environment):
        return environment.get('data', [])

    def act(self, perception):
        if "danger" in perception:
            return "run_away"
        elif "food" in perception:
            return "eat"
        else:
            return "explore"

    def learn(self, experience):
        self.knowledge_base.append(experience)
        return f"Learned: {experience}"

# Contoh penggunaan
ai_agent = SimpleAI("AI-1", ["fire is hot", "water is wet"])
env_data = {"data": ["food", "danger"]}
action = ai_agent.act(ai_agent.perceive(env_data))
print(f"AI Action: {action}")
```

```python
# LOGIKA PROPOSISI DAN RESOLUSI
# Logika Proposisi
class Proposition:
    def __init__(self, symbol, value=None):
        self.symbol = symbol
        self.value = value

    def __str__(self):
        return f"{self.symbol} = {self.value}"

# Contoh proposisi
p = Proposition("P", True)  # P = True
q = Proposition("Q", False) # Q = False

# Operasi logika
def logical_and(p, q):
    return p.value and q.value

def logical_or(p, q):
    return p.value or q.value

def logical_not(p):
    return not p.value

print(f"P AND Q: {logical_and(p, q)}")
print(f"P OR Q: {logical_or(p, q)}")
print(f"NOT P: {logical_not(p)}")

# Resolusi dalam bentuk klausa
def resolve(clause1, clause2):
    # Resolusi sederhana: menghapus kontradiksi
    for literal in clause1:
        if f"~{literal}" in clause2:
            result = [l for l in clause1 if l != literal] + \
                     [l for l in clause2 if l != f"~{literal}"]
            return list(set(result))
    return None

# Contoh klausa
c1 = ["P", "Q", "~R"]
c2 = ["~P", "R", "S"]
resolvent = resolve(c1, c2)
print(f"Resolusi {c1} dan {c2}: {resolvent}")
```

```python
# LOGIKA PREDIKAT DAN KONVERSI
# Predikat sederhana
class Predicate:
    def __init__(self, name, *args):
        self.name = name
        self.args = args

    def __str__(self):
        args_str = ', '.join(str(arg) for arg in self.args)
        return f"{self.name}({args_str})"

# Contoh predikat
likes = Predicate("Likes", "John", "Mary")
student = Predicate("Student", "John")

print(f"Predikat 1: {likes}")
print(f"Predikat 2: {student}")

# Konversi ke bentuk klausa
def to_clausal_form(predicate):
    # Sederhana: predikat menjadi literal positif
    return [str(predicate)]

clause = to_clausal_form(likes)
print(f"Bentuk klausa: {clause}")
```

```python
# POHON KEPUTUSAN
class DecisionNode:
    def __init__(self, attribute=None, value=None, results=None, branches=None):
        self.attribute = attribute  # Atribut yang diuji
        self.value = value          # Nilai atribut
        self.results = results      # Hasil (untuk leaf node)
        self.branches = branches or {}  # Cabang-cabang

    def predict(self, sample):
        if self.results is not None:
            return self.results

        attr_value = sample.get(self.attribute)
        if attr_value in self.branches:
            return self.branches[attr_value].predict(sample)
        return "Unknown"

# Contoh pohon keputusan sederhana
weather_tree = DecisionNode(
    attribute="outlook",
    branches={
        "sunny": DecisionNode(
            attribute="humidity",
            branches={
                "high": DecisionNode(results="don't_play"),
                "normal": DecisionNode(results="play")
            }
        ),
        "rainy": DecisionNode(
            attribute="windy",
            branches={
                "true": DecisionNode(results="don't_play"),
                "false": DecisionNode(results="play")
            }
        )
    }
)

# Prediksi
sample1 = {"outlook": "sunny", "humidity": "high"}
sample2 = {"outlook": "rainy", "windy": "false"}
print(f"Prediksi sample1: {weather_tree.predict(sample1)}")
print(f"Prediksi sample2: {weather_tree.predict(sample2)}")
```

```python
# JARINGAN SEMANTIK
class SemanticNetwork:
    def __init__(self):
        self.nodes = {}
        self.relations = []

    def add_node(self, name, type):
        self.nodes[name] = type

    def add_relation(self, node1, relation, node2):
        self.relations.append((node1, relation, node2))

    def query(self, node1=None, relation=None, node2=None):
        results = []
        for n1, rel, n2 in self.relations:
            if ((node1 is None or node1 == n1) and
                (relation is None or relation == rel) and
                (node2 is None or node2 == n2)):
                results.append((n1, rel, n2))
        return results

# Contoh jaringan semantik
net = SemanticNetwork()
net.add_node("John", "Person")
net.add_node("Prabowo", "Person")
net.add_node("Sawit", "Fruit")
net.add_node("Mary", "Person")
net.add_node("Apple", "Fruit")
net.add_relation("John", "likes", "Mary")
net.add_relation("John", "eats", "Apple")
net.add_relation("Mary", "hates", "Apple")
net.add_relation("Prabowo", "likes", "Sawit")

print(f"Nodes: {net.nodes}")
print(f"Semua relasi: {net.query()}")
print(f"John suka siapa? {net.query(node1='John', relation='likes')}")
print(f"Siapa yang benci apel? {net.query(node1='Mary', relation='hates')}")
print(f"Prabowo suka apa? {net.query(node1='Prabowo', relation='likes')}")
```

```python
# FRAME-BASED REPRESENTATION
class Frame:
    def __init__(self, name, parent=None):
        self.name = name
        self.parent = parent
        self.slots = {}

    def set_slot(self, slot_name, value):
        self.slots[slot_name] = value

    def get_slot(self, slot_name):
        if slot_name in self.slots:
            return self.slots[slot_name]
        elif self.parent:
            return self.parent.get_slot(slot_name)
        return None

    def __str__(self):
        slots_str = ', '.join(f"{k}: {v}" for k, v in self.slots.items())
        return f"Frame({self.name})[{slots_str}]"

# Contoh frame
animal_frame = Frame("Animal")
animal_frame.set_slot("breaths", True)
animal_frame.set_slot("moves", True)

bird_frame = Frame("Bird", animal_frame)
bird_frame.set_slot("has_wings", True)
bird_frame.set_slot("can_fly", True)

penguin_frame = Frame("Penguin", bird_frame)
penguin_frame.set_slot("can_fly", False)
penguin_frame.set_slot("habitat", "Antarctica")

print(f"Animal frame: {animal_frame}")
print(f"Bird frame: {bird_frame}")
print(f"Penguin frame: {penguin_frame}")
print(f"Penguin breaths? {penguin_frame.get_slot('breaths')}")
print(f"Penguin can fly? {penguin_frame.get_slot('can_fly')}")
```

```python
# SCRIPT-BASED REPRESENTATION
class Script:
    def __init__(self, name, roles, props, scenes):
        self.name = name
        self.roles = roles
        self.props = props
        self.scenes = scenes

    def execute_scene(self, scene_name, context):
        if scene_name in self.scenes:
            scene = self.scenes[scene_name]
            result = []
            for action in scene:
                # Ganti placeholder dengan nilai konteks
                executed_action = action
                for key, value in context.items():
                    executed_action = executed_action.replace(f"${key}", str(value))
                result.append(executed_action)
            return result
        return []

    def predict_next(self, current_scene):
        scenes_list = list(self.scenes.keys())
        current_index = scenes_list.index(current_scene)
        if current_index + 1 < len(scenes_list):
            return scenes_list[current_index + 1]
        return None

# Contoh naskah restoran
restaurant_script = Script(
    name="RestaurantVisit",
    roles=["Customer", "Waiter", "Chef"],
    props=["Menu", "Food", "Bill"],
    scenes={
        "Enter": ["$Customer enters restaurant", "$Waiter greets $Customer"],
        "Order": ["$Waiter gives $Menu to $Customer", "$Customer orders $Food"],
        "Eat": ["$Chef prepares $Food", "$Waiter serves $Food", "$Customer eats $Food"],
        "Pay": ["$Waiter brings $Bill", "$Customer pays $Bill"],
        "Exit": ["$Customer leaves restaurant"]
    }
)

# Eksekusi naskah
context = {"Customer": "John", "Waiter": "Mary", "Chef": "Bob",
           "Food": "steak", "Menu": "menu", "Bill": "bill"}
scene_result = restaurant_script.execute_scene("Order", context)
print("Scene 'Order':")
for action in scene_result:
    print(f"  - {action}")

next_scene = restaurant_script.predict_next("Order")
print(f"Scene berikutnya: {next_scene}")
```

```python
# CONTOH
def simple_expert_system():
    """Sistem pakar sederhana untuk diagnosis"""

    # Knowledge base
    symptoms_kb = {
        "fever": ["flu", "covid", "dengue"],
        "cough": ["flu", "covid", "asthma"],
        "headache": ["flu", "migraine", "covid"],
        "rash": ["dengue", "allergy"]
    }

    diseases_kb = {
        "flu": {"treatment": "rest and hydration", "severity": "mild"},
        "covid": {"treatment": "isolation and medical care", "severity": "moderate"},
        "dengue": {"treatment": "hospitalization", "severity": "severe"}
    }

    # Inference engine sederhana
    def diagnose(symptoms):
        possible_diseases = {}
        for symptom in symptoms:
            if symptom in symptoms_kb:
                for disease in symptoms_kb[symptom]:
                    possible_diseases[disease] = possible_diseases.get(disease, 0) + 1

        # Cari penyakit dengan gejala terbanyak
        if possible_diseases:
            most_likely = max(possible_diseases, key=possible_diseases.get)
            confidence = possible_diseases[most_likely] / len(symptoms)
            return most_likely, confidence
        return None, 0

    # Contoh penggunaan
    patient_symptoms = ["fever", "cough", "headache"]
    diagnosis, confidence = diagnose(patient_symptoms)

    if diagnosis:
        treatment = diseases_kb[diagnosis]["treatment"]
        print(f"Diagnosis: {diagnosis} (confidence: {confidence:.0%})")
        print(f"Treatment: {treatment}")
    else:
        print("No diagnosis found")

# Jalankan contoh
print("\n=== SISTEM PAKAR DIAGNOSIS ===")
simple_expert_system()
```
