```python
import heapq  # Digunakan untuk Priority Queue (Min-Heap) pada open_list

def a_star_search(graph, start, goal, heuristic):
    """Mengimplementasikan algoritma A* Search."""

    # open_list adalah priority queue yang menyimpan tuple: (f_cost, node, path)
    # f_cost: g(n) + h(n)
    # node: node saat ini
    # path: jalur dari start ke node saat ini
    open_list = [(heuristic[start], start, [start])]

    # g_cost_map menyimpan biaya sebenarnya dari start ke setiap node yang telah dikunjungi
    g_cost_map = {start: 0}

    # closed_list melacak node yang telah dievaluasi sepenuhnya
    closed_list = set()

    print(f"Memulai A* Search dari {start} ke {goal}\n")

    while open_list:
        # Ambil node dengan f_cost terendah dari open_list
        f_cost, current_node, current_path = heapq.heappop(open_list)

        print(f"Mengunjungi node: {current_node} (f={f_cost}, g={g_cost_map[current_node]}, h={heuristic[current_node]}) | Jalur: {current_path}")

        # Jika node saat ini adalah tujuan, jalur ditemukan
        if current_node == goal:
            print(f"\nNode tujuan ({goal}) ditemukan!")
            return current_path, g_cost_map[current_node]

        # Jika node sudah ada di closed_list, abaikan
        if current_node in closed_list:
            continue

        # Tambahkan node saat ini ke closed_list
        closed_list.add(current_node)

        # Jelajahi tetangga dari node saat ini
        for neighbor, edge_cost in graph.get(current_node, []):
            # Jika tetangga sudah ada di closed_list, abaikan
            if neighbor in closed_list:
                continue

            # Hitung biaya sebenarnya dari start ke tetangga melalui current_node
            tentative_g_cost = g_cost_map[current_node] + edge_cost

            # Jika menemukan jalur yang lebih murah
            if tentative_g_cost < g_cost_map.get(neighbor, float('inf')):
                g_cost_map[neighbor] = tentative_g_cost

                # Hitung f_cost
                f_cost_neighbor = tentative_g_cost + heuristic.get(neighbor, 0)
                new_path = current_path + [neighbor]

                heapq.heappush(open_list, (f_cost_neighbor, neighbor, new_path))
                print(f"  Menambahkan/Memperbarui tetangga {neighbor} (f={f_cost_neighbor}, g={tentative_g_cost}, h={heuristic.get(neighbor, 0)}) ke open_list | Jalur: {new_path}")

    print(f"\nNode tujuan ({goal}) tidak dapat dijangkau dari {start}.")
    return None, None


# ===============================
# GRAF KASUS: PERGI KE KAMPUS
# ===============================

graph_kampus = {
    'Rumah': [('Jalan A', 2), ('Jalan B', 4)],
    'Jalan A': [('Gang C', 2), ('Gang D', 5)],
    'Jalan B': [('Gang D', 1)],
    'Gang C': [('Kampus', 3)],
    'Gang D': [('Kampus', 2)],
    'Kampus': []
}

# Heuristik (perkiraan jarak ke Kampus)
heuristics_kampus = {
    'Rumah': 7,
    'Jalan A': 6,
    'Jalan B': 3,
    'Gang C': 4,
    'Gang D': 2,
    'Kampus': 0
}

print("=== Contoh A* Search: Pergi ke Kampus ===")
path_kampus, cost_kampus = a_star_search(graph_kampus, 'Rumah', 'Kampus', heuristics_kampus)

if path_kampus:
    print(f"Jalur optimal yang ditemukan oleh A*: {path_kampus}")
    print(f"Total jarak yang ditempuh: {cost_kampus} km")
else:
    print("Tidak ada jalur yang ditemukan.")

print("\n=== Contoh A* Search ke lokasi yang tidak ada ===")
path_no_goal, cost_no_goal = a_star_search(graph_kampus, 'Rumah', 'Mall', heuristics_kampus)

if path_no_goal:
    print(f"Jalur optimal yang ditemukan oleh A*: {path_no_goal}")
    print(f"Total jarak yang ditempuh: {cost_no_goal}")
else:
    print("Tidak ada jalur yang ditemukan.")
```
